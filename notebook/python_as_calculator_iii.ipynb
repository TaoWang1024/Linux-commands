```python
# initialization
import math # for log base-10 function
import pint
ur = pint.UnitRegistry()
Q_ = ur.Quantity # to work with decibels
```

```python
# assign values to the physical quantities
L0 = Q_(138, ur.dB)
r0 = 2.2 * ur.m
r = 3 * ur.kilometer
```

```python
# calculate L, the sound level of the whistle 3km away
L = L0 - Q_(20 * math.log10(r / r0), ur.dB)
L # display that sound level
```

```python
# use format printing to print answer in scientific notation, 
# displaying only 2 significant digits
print(f"{L:1.1E}")
```

```python
The answer to the problem statement, by the way, is: Yes, those residents would be able to hear the whistle.
The sound level of the whistle 3 km away is 75 dB, less than that of a ringing telephone (80 dB), but well above the level of normal
conversation (60 dB).
```
