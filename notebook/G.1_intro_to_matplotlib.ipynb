```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# pyplot plots a list of numbers, given as a single argument, as y values against their indices (x-axis)

plt.plot(list(range(10)))
```

```python
# here we use list comprehension to generate the list of y values
# note that the x- and y-axes are not on the same scale

plt.plot([x**2 for x in range(10)])
```

```python
# if we give pyplot two lists, the first is taken as the x values, and the second as the y values
# the lists must be of the same length

plt.plot([2*x for x in range(10)],[x**2 for x in range(10)])
```

```python
# when the axes scales are unequal, the plot distorts the shape of the curve
# to set the axes to the same scale, we use the plt.axis() method, passing the string 'equal' as argument
# note that matplotlib returns the tuple (xmin, xmax, ymin, ymax)

plt.plot([2*x for x in range(10)],[x**2 for x in range(10)])
plt.axis('equal')
```

```python
# the matplotlib inline magic allows for multiple curves to render on the same plot

plt.plot(list(range(10)))
plt.plot([2*x for x in range(10)])
```

```python
# to render multiple curves on the same plot
# the pyplot hold() method must be passed the boolean argument True
# and once we've declared our plots, we use the pyplot show() method to show the plot

#plt.hold(True)
plt.plot(list(range(10)))
plt.plot([2*x for x in range(10)])
plt.show()
```

```python
# likewise, we can render each curve on its own plot

plt.plot(list(range(10)))
plt.show()
plt.plot([2*x for x in range(10)])
plt.show()
```

```python
# we can also make subplots
# first we create an instance of the plt.figure() object
fig = plt.figure()
# then we add each plot to it as a subplot
# the add_subplot() method argument specifies: num_rows, num_cols, plot_number
ax1 = fig.add_subplot(211) # plot number 1
ax1.plot(list(range(10)))
ax2 = fig.add_subplot(212) # plot number 2
ax2.plot([2*x for x in range(10)])
plt.show()
```

```python
fig = plt.figure()
# then we add each plot to it as a subplot
# the add_subplot() method argument specifies: num_rows, num_cols, plot_number
ax1 = fig.add_subplot(121) # plot number 1
ax1.plot(list(range(10)))
ax2 = fig.add_subplot(122) # plot number 2
ax2.plot([2*x for x in range(10)])
plt.show()
```

```python
# let's return to multiple curves on one plot
# we can specify the colors of our curves
# we can add a grid
# we can label our curves and axes, title our plot, and provide a legend

#plt.hold(True)

plt.plot(list(range(10)), label='x', color='magenta') # label is shown in legend
plt.plot([2**x for x in range(10)], label='x**2', color='cyan') # label is shown in legend

plt.grid(True)

plt.xlabel('indices')
plt.ylabel('values')
plt.title('linear v. quadratic growth')

plt.legend()
plt.show()
```
