spherical -> cartesian coordinates

$x = r * cos(\theta) * cos(\lambda)$

$y = r * sin(\theta)$

$z = r * cos(\theta) * sin(\lambda)$

```python
E.<x,y,z> = EuclideanSpace()
E.atlas()
cartesian = E.cartesian_coordinates()
```

```python
cartesian.coord_range()
```

```python
spherical.<r,la, th> = E.spherical_coordinates()
spherical.coord_range()
```

```python
spherical.plot(cartesian, color={r:'red', th:'green', la:'orange'})
```

the red lines are those along which $r$ varies, while ($\lambda, \theta$) are kept fixed;

the grid lines are those along which $\lambda$ varies, while (r, $\theta$) are kept fixed;

the orange lines are those along which $\theta$ varies, while (r, $\lambda$) are kept fixed.
