```python
# initialization
import pint
ur = pint.UnitRegistry()
```

```python
# we will use multiple assignment statement to create and instantiate the variables
h, A, T = .47, 30 * ur.sq_ft, 15 * ur.delta_degF
```

```python
# heat loss in btu per hour
# use format printing to print answer in scientific notation, 
# displaying only 2 significant digits
q = h * A * T
print(f"{q:1.1E}")
```
