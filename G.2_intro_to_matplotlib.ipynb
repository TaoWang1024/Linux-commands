```python
%matplotlib inline

from IPython.display import YouTubeVideo
import numpy as np
import matplotlib.pyplot as plt
```

```python
# scatter plots and color maps
n = 50
x = np.linspace(0,10, n)
noise = np.random.rand(n) * 5.
y1 = x + noise
```

```python
# the default colormap in matplotlib used to be the same as in matlab. 
# it's called 'jet', and it's infamous for how much it distorts data.

plt.figure()
areas = np.random.rand(n) * 300
colors = np.random.rand(n)
plt.scatter(x, y1, c=colors, s=areas, cmap='jet')
plt.colorbar()
```

```python
# the current default for matplotlib is now 'viridis'
# it's what you get when you don't specify a cmap argument

plt.figure()
areas = np.random.rand(n) * 300
colors = np.random.rand(n)
plt.scatter(x, y1, c=colors, s=areas, cmap='viridis')
plt.colorbar()
```

```python
# grayscale works great for single-dimensional data

figure = plt.figure()
areas = np.random.rand(n) * 300
colors = np.random.rand(n)
plt.scatter(x, y1, c=colors, s=areas, cmap='gray')
plt.colorbar()
plt.show()
```

```python
# image plots

# create some data
e1 = np.random.rand(100)
e2 = np.random.rand(100) * 2
e3 = np.random.rand(100) * 10
e4 = np.random.rand(100) * 100

corrmatrix = np.corrcoef([e1, e2, e3, e4])

plt.imshow(corrmatrix, interpolation='nearest', cmap='rainbow_r')
plt.colorbar()
```
